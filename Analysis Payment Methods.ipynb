```python
df.rename(columns={'YearMonth':'Data', 'quantityPix':'VolumePix', 'valuePix':'ValorPix','quantityTED':'VolumeTED', 'valueTED': 'ValorTED',
                 'quantityTEC':'VolumeTEC','valueTEC':'ValorTEC', 'quantityBankCheck':'VolumeCheque', 'valueBankCheck':'ValorCheque', 
                 'quantityBrazilianBoletoPayment':'VolumeBoleto', 'valueBrazilianBoletoPayment': 'ValorBoleto','quantityDOC':
                 'VolumeDOC', 'valueDOC':'ValorDOC'}, inplace=True)
df.head()
```

```python
df['Data'] = pd.to_datetime(df['Data'], format='%Y%m')
df
```

```python
metodos_pagamento = ['Pix', 'TED','TEC','Cheque','Boleto','DOC']
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(20, 25))
```

```python

for method in metodos_pagamento:
    plt.plot(df['Data'], df[f'Volume{method}'], label=method)

plt.title('TendÃªncia de Quantidade de TransaÃ§Ãµes por MÃ©todo')
plt.xlabel('Data')
plt.ylabel('Quantidade de TransaÃ§Ãµes em BilhÃµes')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
```

```python

for method in metodos_pagamento:
    plt.plot(df['Data'], df[f'Valor{method}'], label=method)

plt.title('TendÃªncia de Valor de TransaÃ§Ãµes por MÃ©todo')
plt.xlabel('Data')
plt.ylabel('Valor de TransaÃ§Ãµes em BilhÃµes (R$)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
```

```python

total_values = {method: df[f'Valor{method}'].sum() for method in metodos_pagamento}

# Preparar dados para o grÃ¡fico
labels = list(total_values.keys())
sizes = list(total_values.values())

# Calcular porcentagens
total = sum(sizes)
percentages = [(value/total) for value in sizes]

# Criar grÃ¡fico de pizza
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=[f'{l}\n{p:.2f}%' for l, p in zip(labels, percentages)], autopct='%1.1f%%')
plt.title('DistribuiÃ§Ã£o do Valor Total por MÃ©todo de Pagamento')

# Adicionar legenda com valores absolutos
legend_labels = [f'{l}: R$ {v:,.2f}' for l, v in total_values.items()]
plt.legend(legend_labels, title='Valores Totais em MilhÃµes de Reais', loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()

# Imprimir valores detalhados
print("Valores totais por mÃ©todo de pagamento em MilhÃµes de Reais:")
for method, value in total_values.items():
    percentage = (value / total) 
    print(f"{method}: R$ {value:,.2f} ({percentage:.2f}%)")
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from statsmodels.tsa.arima.model import ARIMA


df = df.sort_values('Data')


def prepare_lstm_data(data, time_steps=3):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:(i + time_steps)])
        y.append(data[i + time_steps])
    return np.array(X), np.array(y)


def train_lstm_model(X_train, y_train):
    model = Sequential([
        LSTM(50, activation='relu', input_shape=(X_train.shape[1], 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=50, verbose=0)
    return model


pix_quantity = df['VolumePix']
pix_value = df['ValorPix']
scaler_qty = MinMaxScaler()
pix_quantity_scaled = scaler_qty.fit_transform(pix_quantity.values.reshape(-1, 1))
scaler_val = MinMaxScaler()
pix_value_scaled = scaler_val.fit_transform(pix_value.values.reshape(-1, 1))





X_qty, y_qty = prepare_lstm_data(pix_quantity_scaled)
X_val, y_val = prepare_lstm_data(pix_value_scaled)


split = int(0.8 * len(X_qty))
X_train_qty, X_test_qty = X_qty[:split], X_qty[split:]
y_train_qty, y_test_qty = y_qty[:split], y_qty[split:]

X_train_val, X_test_val = X_val[:split], X_val[split:]
y_train_val, y_test_val = y_val[:split], y_val[split:]


lstm_model_qty = train_lstm_model(X_train_qty, y_train_qty)
lstm_model_val = train_lstm_model(X_train_val, y_train_val)


predictions_qty = lstm_model_qty.predict(X_test_qty)
predictions_val = lstm_model_val.predict(X_test_val)


predictions_qty_inv = scaler_qty.inverse_transform(predictions_qty)
y_test_qty_inv = scaler_qty.inverse_transform(y_test_qty)
predictions_val_inv = scaler_val.inverse_transform(predictions_val)
y_test_val_inv = scaler_val.inverse_transform(y_test_val)


print("MÃ©tricas para Volume de Pix:")
print(f"MSE: {mean_squared_error(y_test_qty_inv, predictions_qty_inv)}")
print(f"MAE: {mean_absolute_error(y_test_qty_inv, predictions_qty_inv)}")

print("\nMÃ©tricas para Valor de Pix:")
print(f"MSE: {mean_squared_error(y_test_val_inv, predictions_val_inv)}")
print(f"MAE: {mean_absolute_error(y_test_val_inv, predictions_val_inv)}")


plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.plot(y_test_qty_inv, label='Real')
plt.plot(predictions_qty_inv, label='Previsto')
plt.title('PrevisÃ£o de Volume de Pix em BilhÃµes')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(y_test_val_inv, label='Real')
plt.plot(predictions_val_inv, label='Previsto')
plt.title('PrevisÃ£o de Valor de Pix em BilhÃµes')
plt.legend()

plt.tight_layout()
plt.show()

# Modelo ARIMA para comparaÃ§Ã£o
def arima_forecast(data, order=(1,1,1)):
    model = ARIMA(data, order=order)
    results = model.fit()
    forecast = results.forecast(steps=5)
    return forecast

# PrevisÃ£o ARIMA
arima_forecast_qty = arima_forecast(pix_quantity)
arima_forecast_val = arima_forecast(pix_value)

print("\nPrevisÃ£o ARIMA para Volume de Pix:")
print(arima_forecast_qty)

print("\nPrevisÃ£o ARIMA para Valor de Pix:")
print(arima_forecast_val)
```
